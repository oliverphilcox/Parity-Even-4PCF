### Load 3PCF

```python
# C++ implementation
all_threepcf = []
flat_threepcf = []
for sim_no in range(1,1201):
    infile = indir+'lognormal%d.zeta_3pcf.txt'%sim_no
    if not os.path.exists(infile): continue
    bins1,bins2 = np.asarray(np.loadtxt(infile,skiprows=3,max_rows=2),dtype=int)
    # Define product of three radial bins
    r1_3pcf = r_from_bin(bins1)
    r2_3pcf = r_from_bin(bins2)
    ell1 = np.asarray(np.loadtxt(infile,skiprows=8)[:,:1],dtype=int).T[0]
    good_ell = (ell1<=l_max)
    ell1 = ell1.copy()[good_ell]
    indat = np.loadtxt(infile,skiprows=8)[:,1:][good_ell]
    all_threepcf.append(indat)
    flat_threepcf.append((indat).ravel())
flat_threepcf = np.asarray(flat_threepcf)
n_mocks = len(all_threepcf)
n_radial = len(bins1)
n_mult = len(ell1)
print("Using %d mocks"%n_mocks)
threepcf_mean = np.mean(all_threepcf,axis=0)
threepcf_std = np.std(all_threepcf,axis=0)#/np.sqrt(n_mocks)
```

### Compute Theory Model

##### Define bin-integrated spherical Bessel functions

```python
from scipy.interpolate import interp1d
from scipy.special import spherical_jn
from scipy.integrate import simps
from sympy.physics.wigner import wigner_3j
from scipy.special import sici

## Define bin-integrated Bessel functions. These are Integral[r^2 j_ell(kr)]/Integral[r^2] which are analytic.
## sici[0] is the Sine integral.
## ell>5 can be easily implemented also if needed

def integ_bessel(ell,bin1):
    r_min = bin1*(R_max-R_min)/n_r+R_min
    r_max = (bin1+1.)*(R_max-R_min)/n_r+R_min
    if ell==0:
        tmp1 = (-k*r_max*np.cos(k*r_max)+np.sin(k*r_max))/k**3. 
        tmp2 = (-k*r_min*np.cos(k*r_min)+np.sin(k*r_min))/k**3.
    elif ell==1:
        tmp1 = -2.*np.cos(k*r_max)/k**3. - r_max*np.sin(k*r_max)/k**2.
        tmp2 = -2.*np.cos(k*r_min)/k**3. - r_min*np.sin(k*r_min)/k**2.
    elif ell==2:
        tmp1 = (r_max*np.cos(k*r_max))/k**2 - (4*np.sin(k*r_max))/k**3 + (3*sici(k*r_max)[0])/k**3
        tmp2 = (r_min*np.cos(k*r_min))/k**2 - (4*np.sin(k*r_min))/k**3 + (3*sici(k*r_min)[0])/k**3
    elif ell==3:
        tmp1 = (7.*np.cos(k*r_max))/k**3 - (15.*np.sin(k*r_max))/(k**4*r_max) + (r_max*np.sin(k*r_max))/k**2
        # avoid zero errors!
        if r_min==0:
            tmp2 = -(8./k**3.)
        else:
            tmp2 = (7.*np.cos(k*r_min))/k**3 - (15.*np.sin(k*r_min))/(k**4*r_min) + (r_min*np.sin(k*r_min))/k**2
    elif ell==4:
        tmp1 = (105.*np.cos(k*r_max))/(2.*k**4*r_max) - (r_max*np.cos(k*r_max))/k**2 + (11*np.sin(k*r_max))/k**3 -(105*np.sin(k*r_max))/(2.*k**5*r_max**2) + (15*sici(k*r_max)[0])/(2.*k**3)
        if r_min==0:
            tmp2 = 0.
        else:
            tmp2 = (105.*np.cos(k*r_min))/(2.*k**4*r_min) - (r_min*np.cos(k*r_min))/k**2 + (11*np.sin(k*r_min))/k**3 -(105*np.sin(k*r_min))/(2.*k**5*r_min**2) + (15*sici(k*r_min)[0])/(2.*k**3)
    elif ell==5:
        tmp1 = ((315*k*r_max - 16*k**3*r_max**3)*np.cos(k*r_max) - (315 - 105*k**2*r_max**2 + k**4*r_max**4)*np.sin(k*r_max))/(k**6*r_max**3)
        if r_min==0:
            tmp2 = -16./k**3.
        else:
            tmp2 = ((315*k*r_min - 16*k**3*r_min**3)*np.cos(k*r_min) - (315 - 105*k**2*r_min**2 + k**4*r_min**4)*np.sin(k*r_min))/(k**6*r_min**3)
    else:
        raise Exception("not implemented yet!")
    return (tmp1-tmp2)/((r_max**3.-r_min**3.)/3.)
```

##### Load Pk theory from file

```python
# k,pk0_in,pk2_in,pk4_in = np.loadtxt(indir+'mocks/input_pk_lognormal_real.txt').T
# Pk_mult = [pk0_in,pk2_in,pk4_in]

# Define volume and shot-noise
vol = 3.9e9
Ngal = 587071
ngal = Ngal*1./vol
Pshot = 1./ngal

# reduce k dimension by factor `reduction' for speed
reduction = 2
print("Reducing k-size by %dx"%reduction)
k,pk0_in = np.loadtxt(indir+'mocks/input_pk_lognormal_real.txt').T[:,::reduction]
pk2_in = np.zeros_like(pk0_in)
pk4_in = np.zeros_like(pk2_in)
Pk_mult = [pk0_in+Pshot,pk2_in,pk4_in]

from mcfit import P2xi
r,xi0_in = P2xi(k)(pk0_in)

s_array = np.arange(0.5,1000,0.5) # array of s values
xi0_array = interp1d(r,xi0_in)(s_array+1e-3)
```

```python
len(s_array),len(k)
```

```python
# Compute 2PCF
single_f0L = lambda L,bin1: (-1.)**(L/2)*simps(k**2.*Pk_mult[L//2]*integ_bessel(L,bin1),k)/(2.*np.pi**2.)
twopcf_theory = [np.asarray([single_f0L(2*L,i) for i in range(10)]) for L in range(3)]
```

##### Theory 3PCF

```python
jn_arr = np.asarray([spherical_jn(ell,k[:,np.newaxis]*s_array[np.newaxis,:]) for ell in range(l_max*2+1)])
```

```python
flll_arr = lambda ell,ell1,ell2,bin1,bin2: simps((k**2.*Pk_mult[0]*integ_bessel(ell1,bin1)*integ_bessel(ell2,bin2))[:,np.newaxis]*jn_arr[ell],k,axis=0)/(2.*np.pi**2.)
fll_arr = lambda ell,ell1,bin1: simps((k**2.*Pk_mult[0]*integ_bessel(ell1,bin1))[:,np.newaxis]*jn_arr[ell],k,axis=0)/(2.*np.pi**2.)
```

```python
# Compute matrix of all necessary f_{L,l1}(s,a) kernels
print("Computing all relevant f_{L,L}^{a} kernels")
all_fll = np.zeros((l_max+1,n_r,len(s_array)))
for L in range(0,l_max+1):
    for a in range(n_r):
        all_fll[L,a] = fll_arr(L,L,a)
            
# Compute matrix of all necessary f_{L,l1,l2}(s,a,b) kernels
print("Computing all relevant f_{L,l1,l2}^{a,b} kernels")
all_flll = np.zeros((2*l_max+1,l_max+1,l_max+1,n_r,n_r,len(s_array)))
for L in range(0,2*l_max+1):
    print("L = %d"%L)
    for l1 in range(0,l_max+1):
        for l2 in range(0,l_max+1):
            # check 3j
            if wigner_3j(L,l1,l2,0,0,0)==0: continue
            for a in range(n_r):
                for b in range(n_r):
                    all_flll[L,l1,l2,a,b] = flll_arr(L,l1,l2,a,b)
```

##### Accumulate 3PCF

```python
cov_I = np.zeros((n_radial*n_mult,n_radial*n_mult))
cov_II = np.zeros((n_radial*n_mult,n_radial*n_mult))

for l in range(l_max+1):
    print("Using ell = %d"%l)
    for lp in range(l_max+1):
        print("Using ell-prime = %d"%lp)
        for L in range(2*l_max+1): # range must be up to maximum allowed value from 3j symbols!
            wig = wigner_3j(l,lp,L,0,0,0)
            if wig==0: continue
                
            fact = (4.*np.pi)**3.*wig**2.*(2.*L+1.)*np.sqrt((2.*l+1.)*(2.*lp+1.))/vol
            phase_I = pow(-1.,L)
            phase_II = pow(-1.,(L+l+lp)/2)
                
            i = 0
            for a in range(n_r):
                for b in range(a+1,n_r):
                    assert a==bins1[i], b==bins2[i]
                    
                    iP = 0
                    for ap in range(n_r):
                        for bp in range(ap+1,n_r):
                            assert ap==bins1[iP], bp==bins2[iP]

                            ### First integral contribution
                            kernel_I = all_flll[L,l,lp,a,ap]*all_flll[L,l,lp,b,bp]
                            kernel_I += all_flll[L,l,lp,b,ap]*all_flll[L,l,lp,a,bp]
                            cov_I[l*n_radial+i,lp*n_radial+iP] += fact*phase_I*simps(s_array**2*xi0_array*kernel_I,s_array)

                            ### Second integral contribution
                            kernel_II = all_fll[l,a]*all_fll[lp,ap]*all_flll[L,l,lp,b,bp]
                            kernel_II += all_fll[l,b]*all_fll[lp,ap]*all_flll[L,l,lp,a,bp]
                            kernel_II += all_fll[l,a]*all_fll[lp,bp]*all_flll[L,l,lp,b,ap]
                            kernel_II += all_fll[l,b]*all_fll[lp,bp]*all_flll[L,l,lp,a,ap]

                            cov_II[l*n_radial+i,lp*n_radial+iP] += fact*phase_II*simps(s_array**2*kernel_II,s_array)

                            iP += 1                              
                    i += 1
            
cov_3pcf_theory = cov_I+cov_II
corr_3pcf_theory = np.zeros_like(cov_3pcf_theory)
for i in range(len(cov_I)):
    for j in range(len(cov_II)):
        corr_3pcf_theory[i,j] = cov_3pcf_theory[i,j]/np.sqrt(cov_3pcf_theory[i,i]*cov_3pcf_theory[j,j])
```

##### Theory 4PCF

```python
# Define f^L_{0,l1,l2} kernels as integral over P_L(k)
simple_fL = lambda L,ell1,ell2,bin1,bin2: simps(k**2.*Pk_mult[L//2]*integ_bessel(ell1,bin1)*integ_bessel(ell2,bin2),k)/(2.*np.pi**2.)

# Compute matrix of all necessary f^L_{l1,l2}(a,b) = f^L_{0,l1,l2}(0,a,b) kernels
print("Computing all relevant f^L_{l1,l2} kernels")
all_f = np.zeros((3,l_max+1,l_max+1,n_r,n_r))
for L in [0,2,4]:
    for l1 in range(0,l_max+1):
        for l2 in range(0,l_max+1):
            for a in range(n_r):
                for b in range(n_r):
                    all_f[L//2,l1,l2,a,b] = simple_fL(L,l1,l2,a,b)
```

```python
# Now compute full Gaussian integral
print("Assembling theory model")
fourpcf_theory = np.zeros_like(fourpcf_mean)
for i in range(len(fourpcf_theory)):
    l1,l2,l3 = ell1[i],ell2[i],ell3[i]
    for j in range(n_radial):
        a = bins1[j]
        b = bins2[j]
        c = bins3[j]
        
        if l1 in [0,2,4]:
            # compute contribution if relevant
            threej = np.float64(wigner_3j(l1,l2,l3,0,0,0))
            if threej==0: continue
            fourpcf_theory[i,j] += threej*(4.*np.pi)**(1.5)*np.sqrt((2.*l2+1)*(2.*l3+1.)/(2.*l1+1.)**3.)*(-1.)**((l2-l3)/2.)*twopcf_theory[l1//2][a]*all_f[l1//2,l2,l3,b,c]
        
        if l2 in [0,2,4]:
            # compute contribution if relevant
            threej = np.float64(wigner_3j(l1,l2,l3,0,0,0))
            if threej==0: continue
            fourpcf_theory[i,j] += threej*(4.*np.pi)**(1.5)*np.sqrt((2.*l1+1)*(2.*l3+1.)/(2.*l2+1.)**3.)*(-1.)**((l1-l3)/2.)*twopcf_theory[l2//2][b]*all_f[l2//2,l1,l3,a,c]
        
        if l3 in [0,2,4]:
            # compute contribution if relevant
            threej = np.float64(wigner_3j(l1,l2,l3,0,0,0))
            if threej==0: continue
            fourpcf_theory[i,j] += threej*(4.*np.pi)**(1.5)*np.sqrt((2.*l1+1)*(2.*l2+1.)/(2.*l3+1.)**3.)*(-1.)**((l1-l2)/2.)*twopcf_theory[l3//2][c]*all_f[l3//2,l1,l2,a,b]
    
```
